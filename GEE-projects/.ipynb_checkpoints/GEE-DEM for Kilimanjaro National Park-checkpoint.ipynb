#### Digital Elevation Model

```python
## load python package for analysis
```

```python
import ee
import geemap
import geemap.colormaps as cm
```

```python
## Draw a map
```

```python
Map = geemap.Map(center = (-3.08,37.36), zoom = 11)
Map
```

```python
## load DEM data
```

```python
dem = ee.Image("CGIAR/SRTM90_V4")
## Access dem palettes
palettes = cm.palettes.dem
## Visualize palletes
viz_pal = {'min':0, 'max':6000,'palette':palettes}
## Add dem data as layer to map
Map.addLayer(dem,viz_pal, 'Mt Kilimanjaro DEM')
Map.add_colorbar(viz_pal, orientation = 'horizontal', label = 'Elevation (m)',layer_name = 'DAR DEM')
```
