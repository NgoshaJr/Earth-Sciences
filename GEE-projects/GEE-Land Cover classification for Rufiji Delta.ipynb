```python
## Land Cover classification- for Rufiji Delta in Tanzania using Copernicus Global Land sevices 2019 at 100 m resolution
```

```python
import ee
import geemap
```

```python
Map = geemap.Map(center = (-7.83,39.07), zoom = 10)
Map.add_basemap('HYBRID')
Map
```

```python
## load landcover data
legend_dict = {
    'Open Water': '466b9f',
    'Open Space': 'dec5c5',
    'Developed, Low Intensity': 'eb0000',
    'Deciduous Forest': '68ab5f',
    'Evergreen Forest': '1c5f2c',
    'Pasture': 'dcd939',
    'Wetlands': 'b8d9eb',
}
landcover = ee.Image("COPERNICUS/Landcover/100m/Proba-V-C3/Global/2019").select('discrete_classification')
Map.addLayer(landcover,{},'Pwani Landcover')
Map.add_legend(title="NLCD Land Cover Classification", legend_dict=legend_dict, layer_name='NLCD Land Cover',legend_title='Land cover classes')
```
